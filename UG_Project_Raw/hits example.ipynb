```python
def demo(events, event_number, idx_track, idx_shower):
    
    w_hits = np.concatenate(events.filter_by_event(events.reco_hits_w, event_number))
    x_hits = np.concatenate(events.filter_by_event(events.reco_hits_x_w, event_number))
    
    w_shower = events.filter_by_event(events.reco_hits_w, event_number)[idx_shower]
    x_shower = events.filter_by_event(events.reco_hits_x_w, event_number)[idx_shower]
    
    w_track = events.filter_by_event(events.reco_hits_w, event_number)[idx_track]
    x_track = events.filter_by_event(events.reco_hits_x_w, event_number)[idx_track]
    
    plt.figure(figsize = (8,6))
    plt.scatter(w_hits, x_hits, c='grey', s=3, label='Hits')
    plt.scatter(w_track, x_track, c='cyan', s=3, label='Track')
    plt.scatter(w_shower, x_shower, c='magenta', s=3, label='Shower')
    plt.title(f'Event display for event {event_number}')
    plt.ylabel('X (W View)')
    plt.xlabel('W (Wire Position)')
    plt.legend()
    plt.grid(True)
    plt.savefig("demo_event", bbox_inches='tight', dpi=300)
    plt.show()
```

```python
demo(cheated_0, 65, 1, 10)
```
